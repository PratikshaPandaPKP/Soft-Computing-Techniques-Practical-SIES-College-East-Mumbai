```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import RBFInterpolator

# Generate sample data
x = np.array([0, 1, 2, 3, 4, 5])
y = np.sin(x)  # Sample function values

# Create grid for interpolation
xi = np.linspace(0, 5, 100)
yi = np.sin(xi)  # Ground truth for comparison

# Create RBF Interpolator
rbf = RBFInterpolator(x[:, np.newaxis], y, kernel='linear')

# Interpolate
yi_rbf = rbf(xi[:, np.newaxis])

# Plotting results
plt.figure(figsize=(10, 6))
plt.plot(x, y, 'ro', label='Sample Data (sin(x))')
plt.plot(xi, yi, 'b-', label='True Function (sin(x))')
plt.plot(xi, yi_rbf, 'g--', label='RBF Interpolated')
plt.title('Radial Basis Function Interpolation')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.grid()
plt.show()
```
